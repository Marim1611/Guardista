```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
names = []
for i in range(136):
    names.append('X'+str(i))
names.append('y')
names
```

```python
data_122 = pd.read_csv('experiments/tokenize_only/files_representations_122.csv', names=names)
data_122.head()
```

```python
data_122_safe = pd.read_csv('experiments/tokenize_only/files_representations_122_safe.csv', names=names)
data_122_safe.head()
```

```python
# total = pd.concat([data_78, data_78_safe[0:600]])
total = pd.concat([data_122, data_122_safe])
total
```

```python
def X_y_split(df):
    df_numpy = df.to_numpy()
    features = df_numpy[:, :-1]
    labels = df_numpy[:, -1]
    # or without converting to numpy
    # X = df.iloc[:, :-1]
    # y = df.iloc[:, -1]
    X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=0)
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = X_y_split(total)
X_train, X_test, y_train, y_test
```

```python
%%time
# We use time to compute the training time of our model
model = LogisticRegression(random_state=0, penalty="none").fit(X_train, y_train)
```

```python
y_train_predict = model.predict(X_train)
print(f"Training Accurracy: {accuracy_score(y_train, y_train_predict) * 100}%")
y_test_predict = model.predict(X_test)
print(f"Validation Accurracy: {accuracy_score(y_test, y_test_predict) * 100}%")
print(classification_report(y_test, y_test_predict))
```

```python
import pickle
pickle.dump(model, open('LR_binary_122.pkl', 'wb'))
```

```python
%%time
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
```

```python
y_train_predict = clf.predict(X_train)
print(f"Training Accurracy: {accuracy_score(y_train, y_train_predict) * 100}%")
y_test_predict = clf.predict(X_test)
print(f"Testing Accurracy: {accuracy_score(y_test, y_test_predict) * 100}%")
print(classification_report(y_test, y_test_predict))
```
